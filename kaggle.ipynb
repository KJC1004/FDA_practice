# [Reference Link](https://www.kaggle.com/ronitf/heart-disease-uci)

##### Type A
| column | description |
|---|---|
| age | age in years |
| trestbps | resting blood pressure (in mm Hg on admission to the hospital) |
| chol | serum cholestoral in mg/dl |
| thalach | maximum heart rate achieved |
| oldpeak | ST depression induced by exercise relative to rest |

##### Type B
| column | description |
|---|---|
| cp | chest pain type |
| restecg | resting electrocardiographic results |
| slope | the slope of the peak exercise ST segment |
| ca | number of major vessels (0-3) colored by flourosopy |
| thal | 3 = normal; 6 = fixed defect; 7 = reversable defect |

##### Type C
| column | description |
|---|---|
| sex | (1 = male; 0 = female) |
| fbs | (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false) |
| exang | exercise induced angina (1 = yes; 0 = no) |
| target | 1 or 0 |

# Init

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

pd.options.display.max_rows = 10
figsize = (20,10)
```

```python
heart = pd.read_csv('heart.csv')
```

```python
cols_a = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
cols_b = ['cp', 'restecg', 'slope', 'ca', 'thal']
cols_c = ['sex', 'fbs', 'exang', 'target']
```

# coutplot (Bar)

```python
b = random.choice(cols_b)
sns.countplot(x=b, data=heart)
```

# kdeplot

```python
a0, a1 = random.sample(cols_a, 2)
sns.kdeplot(heart[a0], heart[a1])
```

# distplot (Histogram)

```python
b = random.choice(cols_b)
sns.distplot(heart[b], kde=False)
```

# jointplot (Scatterplot and hexplot)

```python
x, y = random.sample(cols_a, 2)
sns.jointplot(x=x, y=y, data=heart, kind='hex')
```

# boxplot

```python
x, y = random.choice(cols_b), random.choice(cols_a)
sns.boxplot(x=x, y=y, data=heart)
```

# violinplot

```python
x, y = random.choice(cols_b), random.choice(cols_a)
sns.violinplot(x=x, y=y, data=heart)
```

# Faceting

```python
col, row = random.sample(cols_c, 2)
index = random.choice(cols_a)

order = [0,1]
g = sns.FacetGrid(heart, row=row, col=col)
g.map(sns.kdeplot, index)
```

# pairplot

```python
sns.pairplot(heart[cols_a], diag_kind='kde', kind='reg')
```

# heatmap

```python
plt.figure(figsize=figsize)
sns.heatmap(heart[cols_a+cols_b+cols_c].corr(), annot=True)
```

# clustermap

```python
sns.clustermap(heart.corr())
```
